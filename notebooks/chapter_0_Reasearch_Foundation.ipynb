# ── ABSTRACT ─────────────────────────────────────────────────

Predicting whether a policyholder will file an insurance claim is one of the most
valuable problems in actuarial analytics. Accurate claim prediction allows insurers
to price risk more fairly, allocate reserves more efficiently, and flag high-risk
policyholders before losses occur.

This project builds and evaluates a machine learning pipeline for binary claim
prediction using the Porto Seguro Safe Driver dataset — a real-world insurance
dataset with over 595,000 records. Three models are trained and compared:
Logistic Regression as an interpretable statistical baseline, Random Forest as a
tree-based ensemble method, and XGBoost as an advanced gradient boosting algorithm.

A key challenge in insurance data is class imbalance — the majority of policyholders
do not file claims. SMOTE was initially considered to address class imbalance. However, due to large dataset size and computational constraints, class-weighted models were used instead. Model performance is measured using AUC-ROC, F1-score, precision and
recall. SHAP values are used to explain which features drive each prediction.
Finally, predicted probabilities are converted into a 0–100 risk scoring system,
categorising policyholders as Low, Medium, or High risk.

**Keywords:** Insurance Analytics, Claim Prediction, Machine Learning, Risk Scoring,
XGBoost, SHAP, Class Imbalance, Risk Analysis.


# ── INTRODUCTION ─────────────────────────────────────────────
## 1. Introduction

### 1.1 Background

Insurance companies collect large volumes of data about their policyholders — vehicle
details, demographics, driving history, and regional information. The core business
problem is simple: which of these policyholders are most likely to file a claim?

Traditionally, actuaries have used generalised linear models (GLMs) to answer this
question. GLMs are well understood, statistically rigorous, and easy to explain to
regulators. However, they have limitations — they assume linear relationships between
features and outcomes, and they can struggle to capture the complex interactions that
exist in real policyholder data.

Machine learning models offer a powerful complement to traditional actuarial methods.
Algorithms like Random Forest and XGBoost can detect non-linear patterns and feature
interactions that GLMs miss, often producing meaningfully better predictions on
large insurance datasets.

### 1.2 Problem Statement

This project addresses three connected problems:

1. **Prediction:** Can machine learning accurately identify which policyholders are
   likely to file a claim, even when claims are rare events in the data?

2. **Explainability:** Can we understand *why* the model assigns a high or low risk
   score to a given policyholder — in a way that satisfies both business users and
   regulators?

3. **Risk Scoring:** Can raw model probabilities be translated into a simple,
   interpretable risk score that underwriters and pricing teams can act on?

### 1.3 Why This Matters

From an actuarial perspective, the consequences of mis-pricing risk are significant.
Underestimating risk leads to under-reserving and potential losses. Overestimating
risk leads to uncompetitive premiums that drive away low-risk customers. A well-
calibrated, explainable claim prediction model directly supports better pricing
decisions.

This project demonstrates how machine learning can be integrated into an actuarial
workflow — not to replace actuarial judgement, but to enhance it with data-driven
insights.

### 1.4 Scope

This study focuses on binary classification — predicting whether a claim will occur
(1) or not (0). The analysis uses structured tabular data and open-source Python
tools, making the full pipeline reproducible. Deep learning approaches are outside
the scope of this study.

# ── LITERATURE REVIEW ─────────────────────────────────────────
## 3. Literature Review

### 3.1 Machine Learning in Actuarial Science

The integration of machine learning into actuarial practice has been an active area
of research over the past decade. Frees et al. (2014) established that while GLMs
remain the foundation of actuarial claim modelling, they increasingly fall short on
large, complex datasets where non-linear feature interactions are present. This has
opened the door for tree-based and ensemble methods in insurance applications.

Wüthrich (2018) explored the use of neural networks for claims reserving and
pricing, noting their predictive power but highlighting the interpretability
challenge — a key concern in regulated industries like insurance. This tension
between accuracy and explainability is central to the design choices made in this
project.

### 3.2 Ensemble Methods: Random Forest and XGBoost

Breiman (2001) introduced Random Forest as a robust ensemble method that reduces
overfitting by averaging predictions across many decision trees. It has since become
widely used in insurance classification tasks due to its stability and built-in
feature importance measures.

Chen and Guestrin (2016) introduced XGBoost, a gradient boosting framework that
builds trees sequentially, with each tree correcting the errors of the previous one.
XGBoost consistently outperforms other algorithms on structured tabular data and
dominated the Porto Seguro Kaggle competition leaderboard — making it a natural
choice for this dataset.

### 3.3 Handling Class Imbalance

Insurance datasets are typically characterized by a heavy skew: claim events are rare compared to non-claim records. This imbalance can bias models toward predicting "no claim," undermining their ability to identify high-risk policyholders.  

While synthetic oversampling techniques such as SMOTE (Chawla et al., 2002) have historically been applied to rebalance datasets, their computational overhead makes them less practical for very large insurance datasets. In such contexts, class-weighted models are often favored. By assigning higher penalties to misclassified minority-class instances, class weighting achieves a similar corrective effect without the need to generate synthetic data.  

This approach has become standard in large-scale insurance modeling, where efficiency and scalability are critical. It ensures that rare but costly claim events remain visible to the model, while keeping training feasible within real-world computational constraints.  


### 3.4 Model Explainability

Lundberg and Lee (2017) introduced SHAP (SHapley Additive exPlanations), based on
cooperative game theory, as a method for explaining individual model predictions.
Unlike simple feature importance scores, SHAP values show both the magnitude and
direction of each feature's contribution to a specific prediction.

In the insurance context, SHAP is particularly valuable because it allows actuaries
and underwriters to explain to a regulator or customer exactly why a policyholder
was assigned a high risk score — supporting compliance with transparency requirements.

### 3.5 Research Gap

Existing studies typically address prediction, imbalance correction, or explainability
in isolation. This project integrates all three into a single reproducible pipeline,
and extends the output into a practical risk scoring system — bridging the gap between
model output and real-world underwriting application.
